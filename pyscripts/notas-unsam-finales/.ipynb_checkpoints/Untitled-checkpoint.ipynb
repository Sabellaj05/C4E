## Porcentajes de aprobados

```python
import numpy as np
import pandas as pd
import pdfplumber
```

```python

with pdfplumber.open('final.pdf') as pdf:
    data = []
    for pages in pdf.pages:
        text = pages.extract_text()
        lines = text.split('\n')
        
        data.extend([line.split(',') for line in lines])

data2 = []
for l in data:
    for line in l:
        data2.append(line.split(' '))
print(data[:5])
data2[:5]
        
```

```python
name = [name[0] for name in data2]
ape = [ap[1] for ap in data2]
apro = [apro[2] for apro in data2]


df2 = pd.DataFrame(data2, columns=['Nombre', 'Apellido', 'Aprobado'])
df2.head()
```

```python
data2
```

```python
aprobados = [a[-1] for a in data2]

apros = pd.Series(aprobados)

mapa = {'SI': 0, 'NO': 1}

notas = np.vectorize(mapa.get)(apros)

notas = notas[1:]
```

```python
notas = pd.Series(notas)

notas.value_counts()

aprobados = notas[notas == 0]
desaprobados = notas[notas != 0]

print("Porcentaje de aprobados {:.2}%".format(aprobados.count() / notas.count()))
print("Porcentaje de desaprobados {:.2}%".format(desaprobados.count() / notas.count()))
```
